```python
import torch
from ultralytics import RTDETR
import matplotlib.pyplot as plt
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)
```

```python
DATA_YAML = "/content/drive/MyDrive/VisDroneDataset/VisDrone2019_5cls.yaml"

RUN_NAME = "rtdetr_l_visdrone_5cls"
EPOCHS   = 40
IMGSZ    = 640        # default YOLO/RT-DETR 640
BATCH    = 12
```

```python
model = RTDETR("rtdetr-l.pt")
```

```python
model.train(
    data=DATA_YAML,
    epochs=EPOCHS,
    imgsz=IMGSZ,
    batch=BATCH,
    device=0 if device == "cuda" else "cpu",
    workers=4,
    name=RUN_NAME,
    # optional: supaya mirip YOLO kamu
    patience=10,        # early stopping
    cos_lr=True,        # cosine lr (default di versi baru)
    amp=True,           # mixed precision kalau GPU mendukung
)
```

```python
metrics = model.val(
    data=DATA_YAML,
    imgsz=IMGSZ,
    split="val",
    device=0 if device == "cuda" else "cpu",
)

# metrics.box.map    -> mAP50-95 (all classes)
# metrics.box.map50  -> mAP50 (all classes)
# metrics.box.maps   -> list AP per class (mAP50-95)

print("\n==== RT-DETR VAL SUMMARY ====")
print(f"mAP50-95 (all) : {metrics.box.map:.3f}")
print(f"mAP50    (all) : {metrics.box.map50:.3f}")

# -----------------------------
# 5. CETAK AP PER KELAS (BIAR MIRIP YOLO)
# -----------------------------
names = model.names  # dict id -> class name
ap_per_class = metrics.box.maps  # list AP50-95 per class (urutan sama dengan names)

print("\nPer-class AP50-95:")
for cls_id, ap in enumerate(ap_per_class):
    cls_name = names.get(cls_id, str(cls_id))
    print(f"{cls_id:2d} | {cls_name:<12s} | AP50-95 = {ap:.3f}")
```
